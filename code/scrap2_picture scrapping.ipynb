```python
# 0
import urllib
import requests
import urllib.request


from bs4 import BeautifulSoup
```

```python
# request and check response
url='https://yts.mx/'
response=requests.get(url)
response
```

```python
# parse
soup=BeautifulSoup(response.text,'html.parser')
#soup
print(soup.prettify)
```

```python
frames = soup.findAll('div',{'class':'browse-movie-wrap col-xs-10 col-sm-5'})
print(frames)
# iterate over list of frames

for frame in frames:
    # 11 figure in 1 frame :find()
    figure=frame.find('figure')
    photo_url=figure.img['src']
    print(photo_url)
    name=photo_url.split('/')
    name=name[-2]+name[-1]
    print(name)
    print()
    urllib.request.urlretrieve(photo_url,name)
    
#photo = browse-movie-wrap col-xs-10 col-sm-5
#frames = soup.findAll('div',{'class':'browse-movie-wrap col-xs-10 col-sm-5'})
#print(frames)
```
