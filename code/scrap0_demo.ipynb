```python
#!pip install html-parser
# 0 import libraries
import requests
from bs4 import BeautifulSoup
import html.parser
```

```python
# 1 Request
url='https://biography.yourdictionary.com/articles/how-did-bruce-lee-die.html' 
response = requests.get(url)
# 2 Check Response
response
# 200 -> OK
```

```python
# parsing
soup=BeautifulSoup(response.text,'html.parser')
#soup
# to get all p tag
soup.p
```

```python
#dont want p tag in above output
paragraph=soup.p.text
print(type(paragraph))
print()
print(paragraph)
```

```python
# add few word to paragraph
updated_para=paragraph+'Hey this is added text'
updated_para
```

```python
# to get all paragraph
paras=soup.findAll('p')
paras
# here 'p' is html tag that implies paragraph 
```

```python
# to get text from avbove paragraph
# Above text is saved in list. So to get text iterate list
for para in paras:
  print(para.text)
```
